### wap to take user input and check the data,if the data is numeric,convert it to number else if the data is alphanumeric then extract only numbers and convert it.

```python
data= input("please,enter some data:")
if data.isnumeric():
    num=int(data)
    print("data converted to number:",num)
elif data.isalnum:
    temp=''
    for char in data:
        if char.isnumeric():
            temp+=char
        if temp:
            num=int(temp)
            print('data converted to number2:',num)
else:
    print('your data does not contain any number')            
```

### write a simple text menu based calculator for doing basic options like addition,subtraction,multiplication.

```python
print('here is boring calculator menu')
print('1.Add number')
print('2.Sub number')
print('3.Mul number')
ch = int(input('Please enter your choice:'))
if ch == 1:
    num1=int(input('Enter first number:'))
    num2=int(input('Enter second number:'))
    print('Sum is:',num1+num2)
elif ch == 2:
    num1=int(input('Enter first number:'))
    num2=int(input('Enter second number:'))
    print('Sub is:',num1-num2)

elif ch == 3:
    num1=int(input('Enter first number:'))
    num2=int(input('Enter second number:'))
    print('mal is:',num1*num2)
else:
    print('invalid choice:')
```
